```python
# read the dataset
df = pd.read_csv("mushroom.csv")
```

```python
# variables in the dataset
df.info()
```

```python
# finding nulls
df.isnull().sum()
```

```python
# renaming column
df = df.rename(columns={"bruises%3F":"bruises"})
df.head()
```

```python
# drop columns that are not used
df.drop(['cap-surface','cap-color', 'bruises', 'odor', 'gill-attachment',
       'gill-spacing', 'gill-size', 'gill-color', 'stalk-shape', 'stalk-root',
       'stalk-surface-above-ring', 'stalk-surface-below-ring',
       'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type','ring-number', 'ring-type', 'spore-print-color',
       'population', 'habitat'], axis = 1, inplace=True)
```

```python
df.head()
```

```python
# new shape 
df.shape
```

```python
# examining categories
df['cap-shape'].value_counts()
```

```python
df['veil-color'].value_counts()
```

```python
df['class'].value_counts()
```

```python
# frequency encoding first category
fq_cap = df.groupby('cap-shape').size()/len(df)
df.loc[:, "{}_freq_enc".format('cap-shape')] = df['cap-shape'].map(fq_cap)
```

```python
df.head()
```

```python
# frequency encoding second category
fq_veil = df.groupby('veil-color').size()/len(df)
df.loc[:, "{}_freq_enc".format('veil-color')] = df['veil-color'].map(fq_veil)
```

```python
df.head()
```

```python
# frequency encoding the third category
fq_class = df.groupby('class').size()/len(df)
df.loc[:, "{}_freq_enc".format('class')] = df['class'].map(fq_class)
```
